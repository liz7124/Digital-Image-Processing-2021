```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("sIMG_8253.JPG")
plt.imshow(img)
```

```python
img.shape
```

```python
roi = img[100:600,280:750]
plt.imshow(roi,"gray")
```

```python
gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
gray_blur = cv2.GaussianBlur(gray, (15, 15), 0)
thresh = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV, 11, 1)
plt.imshow(thresh,"gray")
```

```python
kernel = np.ones((3, 3), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE,kernel, iterations=2)
plt.imshow(closing,"gray")
```

```python
cont_img = closing.copy()
contours, hierarchy,_ = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

circles = cv2.HoughCircles(closing, cv2.HOUGH_GRADIENT, 0.1, 100, param1 = 30, param2 = 30, minRadius = 100, maxRadius = 180)
```

```python
if circles is not None:
    circles_rnd = np.uint16(np.around(circles))
    print(circles_rnd.shape)
    print(circles_rnd)
    #np.round(circles[0, :]).astype("int")
    
    count = 1
    for i in circles_rnd[0, :]:
        cv2.circle(roi, (i[0],i[1]),i[2],(255, 0, 150), 5)
        #cv2.circle(roi, (i[0],i[1]),6,(0, 0, 0), 4)
        cv2.putText(roi, str(count), (i[0]-30,i[1]+30), cv2.FONT_HERSHEY_PLAIN, 5, (0,0,150), 5)
        count += 1

        plt.rcParams["figure.figsize"] = (4,9)
        plt.imshow(roi)
else:
    print("none")
```

```python
img = cv2.imread("sIMG_8256.JPG")
plt.imshow(img)
```

```python
roi = img[50:750,150:780]
plt.imshow(roi,"gray")
```

```python
gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
gray = cv2.bilateralFilter(gray,23,30,30)
gray = cv2.GaussianBlur(gray, (13,3), 0)
gray_blur = cv2.medianBlur(gray,7)

plt.imshow(gray_blur, 'gray')
```

```python
circles = cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, 0.1, 120, param1 = 50, param2 = 30, minRadius = 5, maxRadius = 200)

if circles is not None:
    circles_rnd = np.uint16(np.around(circles))
    print(circles_rnd.shape)
    print(circles_rnd)
    #np.round(circles[0, :]).astype("int")
    
    count = 1
    for i in circles_rnd[0, :]:
        cv2.circle(roi, (i[0],i[1]),i[2],(255, 0, 150), 5)
        #cv2.circle(roi, (i[0],i[1]),6,(0, 0, 0), 4)
        cv2.putText(roi, str(count), (i[0]-50,i[1]+30), cv2.FONT_HERSHEY_PLAIN, 5, (0,0,150), 5)
        count += 1

        plt.rcParams["figure.figsize"] = (4,9)
        plt.imshow(roi)
else:
    print("none")
```
