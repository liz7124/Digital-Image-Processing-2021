```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.png')
```

```python
img2 = cv2.pyrDown(img)
img3 = cv2.pyrDown(img2)
img4 = cv2.pyrDown(img3)
img5 = cv2.pyrDown(img4)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(151),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(152),plt.imshow(img2)
plt.title('PyrDown 1'), plt.xticks([]), plt.yticks([])
plt.subplot(153),plt.imshow(img3)
plt.title('PyrDown 2'), plt.xticks([]), plt.yticks([])
plt.subplot(154),plt.imshow(img4)
plt.title('PyrDown 3'), plt.xticks([]), plt.yticks([])
plt.subplot(155),plt.imshow(img5)
plt.title('PyrDown 4'), plt.xticks([]), plt.yticks([])
```

```python
img6 = cv2.pyrUp(img5)
img7 = cv2.pyrUp(img6)
img8 = cv2.pyrUp(img7)
img9 = cv2.pyrUp(img8)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(151),plt.imshow(img5)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(152),plt.imshow(img6)
plt.title('PyrUp 1'), plt.xticks([]), plt.yticks([])
plt.subplot(153),plt.imshow(img7)
plt.title('PyrUp 2'), plt.xticks([]), plt.yticks([])
plt.subplot(154),plt.imshow(img8)
plt.title('PyrUp 3'), plt.xticks([]), plt.yticks([])
plt.subplot(155),plt.imshow(img9)
plt.title('PyrUp 4'), plt.xticks([]), plt.yticks([])
```
