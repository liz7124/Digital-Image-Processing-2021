# 1.4.5 Morphological Transformations

## 1. Erosion

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('yes.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
```

## 2. Dilation

```python
dilation = cv2.dilate(img,kernel,iterations = 1)
```

## 3. Opening

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

## 4. Closing

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

## 5. Morphological Gradient

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(331),plt.imshow(img,'gray'),plt.title('Input')
plt.subplot(332),plt.imshow(erosion,'gray'),plt.title('Kernel 5x5')
plt.subplot(333),plt.imshow(dilation,'gray'),plt.title('Dilation')
plt.subplot(334),plt.imshow(opening,'gray'),plt.title('Opening')
plt.subplot(335),plt.imshow(closing,'gray'),plt.title('Closing')
plt.subplot(336),plt.imshow(gradient,'gray'),plt.title('Morphological Gradient')
plt.show()
```
